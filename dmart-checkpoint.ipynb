# date format

```python
dmartdata['Order Date'] = pd.to_datetime(dmartdata['Order Date'])

dmartdata['Order Date']
```

# extract year from date format

```python
dmartdata['Year'] = dmartdata['Order Date'].dt.year
dmartdata['Year']
```

# we need sum of sales values and sum of profit values for each unique category

```python
grouped_data = dmartdata.groupby(['Category', ]).agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()
grouped_data
```

# change data type of column

```python
grouped_data['Profit'] = grouped_data['Profit'].astype(int)
grouped_data['Profit']
```

```python
grouped_data
```

#










# BAR CHART

```python
x = grouped_data['Category']
bar_width = 0.35 # a reasonable bar width
bar = np.arange(len(x))
bar1 = bar + bar_width # fix: Add a *fixed* offset

plt.figure(figsize=(11,6))
plt.bar(bar, grouped_data['Profit'], bar_width, label='Profit')
plt.bar(bar1, grouped_data['Sales'], bar_width, label='Sales')

plt.xticks(bar + bar_width/2, x) # fix: set x-ticks at center of bar groups
plt.ticklabel_format(axis='y', style='plain')

plt.legend(loc='upper left', fontsize=10)




#line graph
plt.plot(bar, grouped_data['Sales'], color='black', linestyle='--', marker='v', label='Sales Trend', linewidth=3)
plt.plot(bar, grouped_data['Profit'], color='black', linestyle='--', marker='v', label='Profit Trend', linewidth=3)


plt.title('Profit and sales per category', pad = 30, fontweight = 'bold', fontsize = 25)
plt.xticks( fontweight='bold')
plt.yticks(fontweight='bold')
plt.tight_layout()
plt.show()
```

#










# Percentage of particular value in any column is 
# (individual value /sum of all values ) * 100

# PIE CHART

```python
plt.figure(figsize=(8, 8)) # Adjust figure size as needed
plt.pie(grouped_data['Sales'], 
        labels=grouped_data['Category'], 
        autopct='%1.1f%%', # Display percentage with 1 decimal place
        startangle=90,
        textprops={'fontsize': 15 },
      
        )     

 

    
plt.title("Sales Distribution by Category",fontsize=25, pad=40,loc='left',color = 'Chocolate', fontweight = 'bold')  # Add a title
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
